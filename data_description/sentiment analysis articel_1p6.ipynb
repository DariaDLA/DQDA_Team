```python
from textblob_de import TextBlobDE as TextBlob
import nltk
import json
```

```python
nltk.download('punkt')
```

```python
with open('article_1p6.txt') as json_file:
    data = json.load(json_file)
print(data)
```

```python
text = data['text']
print(text)                             # Apostroph nicht richtig dargestellt
```

```python
thread = data['thread']
print(thread)
```

```python
blob_text = TextBlob(text)
blob_text.sentiment
```

```python
blob_thread = TextBlob(str(thread))
blob_thread.sentiment
```

```python
blob_thread.sentences # bei den Kommentaren gibt es auch Zitate von anderen Kommentaren (Konversationen zw. den Kommentierenden)
```

```python
for sentence in blob_thread.sentences:
    print(sentence.sentiment.polarity)
```
